```python
def combine_dataframes(dataframes: list, column_names: list, axis=1):
    """
    Combines a list of dataframes with specified column names along a given axis.

    Parameters:
    - dataframes (list): List of dataframes to be combined.
    - column_names (list): List of column names for each dataframe.
    - axis (int): Axis along which to concatenate the dataframes (default is 1).

    Returns:
    - pd.DataFrame: A single concatenated dataframe with updated column names.
    """
    if len(dataframes) != len(column_names):
        raise ValueError("The number of dataframes must match the number of column names.")
    
    updated_dataframes = []
    for df, col_name in zip(dataframes, column_names):
        updated_df = pd.DataFrame(df).T
        updated_df.columns = [col_name]

        updated_df['index'] = range(len(updated_df))
        #print(updated_df['index'].dtype)
        updated_dataframes.append(updated_df)
    
    return pd.concat(updated_dataframes, axis=axis)
```

```python
def add_summary_row(df, indices=None, slice_start=None, row_name='sum_time'):
    """
    Adds a summary row to the DataFrame by summing specific rows or a slice of rows.

    Parameters:
    - df (pd.DataFrame): The input DataFrame.
    - indices (list, optional): List of specific row indices to include in the sum. Default is None.
    - slice_start (int, optional): Start index for a slice of rows. Default is None.
    - row_name (str): Name for the new row index. Default is 'sum_time'.

    Returns:
    - pd.DataFrame: The updated DataFrame with the summary row appended.
    """
    if indices is not None:
        row_sum = df.iloc[indices].sum(axis=0)
    elif slice_start is not None:
        row_sum = df.iloc[slice_start:].sum(axis=0)
    else:
        raise ValueError("Either 'indices' or 'slice_start' must be provided.")
    
    new_row = pd.DataFrame(row_sum).T
    new_row.index = [row_name]
    
    df = pd.concat([df, new_row])
    return df
```
